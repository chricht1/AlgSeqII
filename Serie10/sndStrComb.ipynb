```python
# calculate number possible RNA second strctures
# L - Sequence Length, M - minimal Hairpin Loop length
# Relaxiation: - every Base can pair with every other
def K(L, M):
    if L <= M+2: return 1

    combinations = K(L-1,M) + K(L-1,M) + K(L-2,M)
    for k in range(M, L-M):
        combinations += K(k, M) + K(L-k, M)
    return combinations
```
