```python
# calculate number possible RNA second structures
# L - Sequence Length, 
# M - minimal Hairpin Loop length
# new_bp - indicates if a bp is formed in the new structure so it forms a new structure different from the calling structure
#          which assumes no bp is formed
# start - make structure with no bps count as structure

# recursive function to get all combinations for subsequence from pos. i to j of length L
# start with (i,j) = (1,L)
# Relaxiation: - every Base can pair with every other
def K(L, M, start = 0):
    #print(f'calling K({L})')
    combinations = 0
    if start:
        #print('combinations = 1!')
        combinations = 1

    if L <= M:
        if start: return 1
        return 0

    for l in range(0, L-M-2+1): # all possible bulge loop combinations have the form of pascal triangle
        combinations += l + 1 # every new structure has a bp added after bulge loop, so it forms a new possible combination
        for _ in range(l+1): # add recursive calls
            combinations += K(L-l-2, M)
    #print(f'combinations = {combinations}')

    if L-M-4 >= 0 :# in the second partition min. one bp must be possible otherwise partitioning would be no multiloop
        for k in range(M, L-M-2):   # partition into 2 substructures (by forming a bp) of size k and L-k to account for multiloops 
                                    # (minimal substructure size is M) 
            #print(f'(partition, k in range({M}, {L-M-2})) +=  K({k}, 1) + K({L-k-2}, 0)')
            combinations += 1 + K(k, M) + K(L-k-2, M)
    #print()
    return combinations



L = 3
M = 1
print(f'K({L}, {M})={K(L,M,start=1)}')
```
