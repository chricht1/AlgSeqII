```python
import matplotlib.pyplot as plt
import numpy as np

def read_counts(fpath_bowtie):
    counts = {}
    with open(fpath_bowtie) as f:
        for line in f:
            if 'alignment' in line or '__' in line:
                continue
            else:
                vals = line.split('\t')
                counts[vals[0]] = int(vals[1])
    return counts
```

```python
fpath_bowtie = '../Data/Saccharomyces_RefGenome_R64/Bowtie/htseq_nonunique_transcr_counts.tsv'
fpath_star = '../Data/Saccharomyces_RefGenome_R64/Star/htseq_nonunique_transcr_counts.tsv'

bowtie_counts = read_counts(fpath_bowtie)
star_counts = read_counts(fpath_star)

dataset = []
for key, val in bowtie_counts.items():
    try:
        dataset.append([bowtie_counts[key], star_counts[key]])
    except KeyError:
        print('keyerror')
        pass
dataset = np.int64(dataset)
#dataset = np.int64([[x,y] for [x,y] in dataset if x<80000])

plt.xlabel('Bowtie')
plt.ylabel('Star')
plt.xscale('log')
plt.yscale('log')
plt.scatter(dataset[:,0], dataset[:,1], marker='.')
```
